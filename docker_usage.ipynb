# Usage of docker to provide RESTful API

1. Build the container

```sh
docker build -t demoland_api .
```

2. Run the container\

```sh
docker run -d --name demoland_api -p 80:80 demoland_api
```

3. The API now lives on http://127.0.0.1:80. You can interact with it in the following way:

````python
import requests
````

The API is expecting an input in the JSON format looking like:

```json
{
    "scenario_json": {
        "E00042852":{"signature_type":5,"use":null,"greenspace":null,"job_types":null},
        "E00042915":{"signature_type":5,"use":null,"greenspace":null,"job_types":null},
        "E00042911":{"signature_type":5,"use":null,"greenspace":null,"job_types":null},
        "E00042865":{"signature_type":5,"use":null,"greenspace":null,"job_types":null}
    }
}
```

See the example how to interact with the API using Python and `requests`.

````python
scenario_json = {
    "scenario_json": {
        "E00042852":{"signature_type":5,"use":None,"greenspace":None,"job_types":None},
        "E00042915":{"signature_type":5,"use":None,"greenspace":None,"job_types":None},
        "E00042911":{"signature_type":5,"use":None,"greenspace":None,"job_types":None},
        "E00042865":{"signature_type":5,"use":None,"greenspace":None,"job_types":None}
    }
}
````

````python
r = requests.post("http://127.0.0.1:8000", json=scenario_json)
````
